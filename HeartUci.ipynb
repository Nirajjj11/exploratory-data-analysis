```python
y = df.loc[:,'target']
print(y)
```

```python
X_encode = pd.get_dummies(df, columns=['sex','cp', 'fbs', 'restecg','exang','slope','ca','thal'])
X_encode.head()
```

```python
# Splitting the dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encode, y, test_size=0.2, random_state=1)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear', random_state=0)
svc.fit(X_train,y_train)
```

```python
# Predict the test data
y_pred_svc = svc = svc.predict(X_test)
```

```python
# Checking the performance of the model

y_pred_svc_df = pd.DataFrame({'Expected': y_test, "Predicted" : y_pred_svc})

y_pred_svc_df.head(10)
```

```python
y_pred_svc_df.shape
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
svc_cm = confusion_matrix(y_test, y_pred_svc)

svc_cm
```

```python
svc_acc = accuracy_score(y_test, y_pred_svc)
svc_acc
```
